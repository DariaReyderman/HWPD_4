```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv("mobile.csv")
```

```python
# 1
df_name = df.copy()

df_name['First name'] = df_name['Customer Name'].apply(lambda name: str(name).split()[0])
display(df_name)
```

```python
# 2
df_company = df['Phone Number'].str[0:3]
df['Company'] = df_company.replace({'050' : 'Pelephone', '051' : 'Golan', '052' : 'Selcom', '053' : 'Motorolla', '054' : 'Partner'})
display(df)
```

```python
# 3
# a
display(df['Company'].value_counts())
print()
# b
display(df['Company'].unique())
print()
# c
display(df['Company'].nunique())
print()
```

```python
# 4
# a
df_user = df.copy()

df_user['User type'] = df_user.apply(lambda row: "Talker" if row['Talked'] > row['Internet'] else "Surfer", axis=1)
display(df_user)

# b
def user_type(talking, internet):
    if talking > internet:
        return "Talker"
    else:
        return "Surfer"

df_user['User type'] = np.vectorize(user_type)(df_user['Talked'], df_user['Internet'])
display(df_user)
```

```python
# 5
display(df.sort_values(by=['Talked'], ascending=True))
display(df.sort_values(by=['Talked' , 'Internet'], ascending=False))
```

```python
# 6
display(df.nlargest( 5 , 'Talked'))
display(df.nsmallest( 3 , 'Internet'))
```
